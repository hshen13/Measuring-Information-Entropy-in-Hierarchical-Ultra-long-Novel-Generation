```python
import json
import csv
from pathlib import Path

def extract_numerical_fields(json_path: Path) -> dict:
    try:
        with open(json_path, "r", encoding="utf-8") as f:
            data = json.load(f)

        llm_scores = data.get("llm_detailed_evaluation", {}).get("scores", {})
        text_a = data.get("llm_detailed_evaluation", {}).get("text_a", {})
        text_b = data.get("llm_detailed_evaluation", {}).get("text_b", {})

        return {
            "book_name": data.get("book_name"),  # 中文书名可保留
            "chapter": data.get("chapter"),
            "cosine_similarity": data.get("cosine_similarity"),
            "bert_score_f1": data.get("bert_score_f1"),
            "bert_score_precision": data.get("bert_score_precision"),
            "bert_score_recall": data.get("bert_score_recall"),
            "semantic_similarity": llm_scores.get("semantic_similarity"),
            "plot_similarity": llm_scores.get("plot_similarity"),
            "character_similarity": llm_scores.get("character_similarity"),
            "background_similarity": llm_scores.get("background_similarity"),
            "style_similarity": llm_scores.get("style_similarity"),
            "text_a_props_count": text_a.get("props_count"),
            "text_b_props_count": text_b.get("props_count"),
            "text_a_characters_count": text_a.get("characters_count"),
            "text_b_characters_count": text_b.get("characters_count"),
            "text_a_scenes_count": text_a.get("scenes_count"),
            "text_b_scenes_count": text_b.get("scenes_count"),
        }
    except Exception as e:
        print(f"❌ Error parsing {json_path.name}: {e}")
        return None

def export_directory_to_csv(json_dir: str, output_csv: str):
    json_files = Path(json_dir).rglob("*.json")
    results = [extract_numerical_fields(f) for f in json_files]
    results = [r for r in results if r]

    if not results:
        print("⚠️ No valid JSON files found.")
        return

    output_path = Path(output_csv)
    output_path.parent.mkdir(parents=True, exist_ok=True)

    with open(output_path, "w", newline="", encoding="utf-8-sig") as csvfile:
        writer = csv.DictWriter(csvfile, fieldnames=list(results[0].keys()))
        writer.writeheader()
        writer.writerows(results)

    print(f"✅ 导出完成：{output_csv}")

# ✅ 示例调用
# export_directory_to_csv(
#     "/content/combined_evaluation_results/50000_novel_5000_50000_generated_gemini_evaluation",
#     "/content/combined_evaluation_results/50000_novel_5000_50000_generated_gemini_evaluation.csv"
# )
```

```python
import numpy as np

def mean_confidence_interval(data, confidence=0.95):
    a = np.array(data.dropna())
    n = len(a)
    m = np.mean(a)
    se = np.std(a, ddof=1) / np.sqrt(n)
    h = se * 1.96  # 95% CI
    return m, m - h, m + h

for setting in df_all["setting_id"].unique():
    scores = df_all[df_all["setting_id"] == setting]["semantic_similarity"]
    m, low, high = mean_confidence_interval(scores)
    print(f"{setting:<6} | 语义相似度: {m:.3f} ± {high - m:.3f} (95% CI)")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, linregress

# 构造 DataFrame（来源于 LaTeX 表格数据）
data = {
    "setting_id": ["K1-0", "K2-1", "K2-2", "K2-3", "K2-4", "K2-5"],
    "R": [None, 0.005, 0.01, 0.001, 0.005, 0.01],
    "cosine": [0.677, 0.661, 0.665, 0.650, 0.649, 0.655],
    "bert_f1": [0.199, 0.189, 0.188, 0.168, 0.179, 0.181],
    "semantic": [0.613, 0.591, 0.578, 0.517, 0.572, 0.549],
    "char_sim": [0.566, 0.474, 0.488, 0.354, 0.452, 0.458],
    "style": [0.611, 0.600, 0.581, 0.538, 0.575, 0.572],
    "char_diff": [6.70, 8.62, 8.70, 9.88, 9.30, 9.15],
    "scene_diff": [4.13, 4.65, 4.63, 4.97, 4.81, 4.87],
    "prop_diff": [4.66, 4.88, 5.16, 5.02, 5.20, 5.22]
}
df = pd.DataFrame(data)

# ✅ 去除 None 的 R 用于回归分析
df_r = df.dropna(subset=["R"])

# 1️⃣ 相关性分析：R vs similarity metrics
print("📌 Correlation between compression ratio (R) and similarity metrics:")
for col in ["cosine", "bert_f1", "semantic", "char_sim", "style"]:
    r, p = pearsonr(df_r["R"], df_r[col])
    print(f"R vs {col:<10}: r = {r:.3f}, p = {p:.4f}")

# 2️⃣ 可视化 R vs Semantic Similarity（主指标）
sns.lmplot(x="R", y="semantic", data=df_r)
plt.title("R vs Semantic Similarity")
plt.grid(True)
plt.tight_layout()
plt.show()

# 3️⃣ 回归模型（线性拟合）
slope, intercept, r_value, p_value, std_err = linregress(df_r["R"], df_r["semantic"])
print(f"\n📈 Linear Regression (R → semantic_similarity):")
print(f"y = {slope:.2f} * R + {intercept:.3f} | R² = {r_value**2:.3f}, p = {p_value:.4f}")

# 4️⃣ 分析结构差 vs 相似度指标（全体）
print("\n📌 Structure difference vs similarity (full settings):")
for diff_col, sim_col in [
    ("char_diff", "char_sim"),
    ("scene_diff", "semantic"),
    ("prop_diff", "semantic")
]:
    r, p = pearsonr(df[diff_col], df[sim_col])
    print(f"{diff_col} vs {sim_col}: r = {r:.3f}, p = {p:.4f}")
```
